La differenza principale tra una lista e un dizionario è come i dati sono indicizzati. <br>
in breve:
Lista → accesso per posizione
Dizionario → accesso per chiave (coppie chiave-valore)

```python
my_dict = {'key1': 'value1', 'key2':'value2'}
```

```python
my_dict
```

```python
my_dict['key1']
```

```python
prices_lookup = {'apple':2.99, 'oranges':1.99}
```

```python
prices_lookup['apple']
```

```python
prices_lookup['oranges']
```
